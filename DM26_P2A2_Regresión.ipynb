#  Parte 2: Regresión Lineal Múltiple

````python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
````

````python
data2 = {
    "Horas_estudio": [1.5, 2.0, 3.0, 3.5, 4.0, 5.5, 6.0, 6.5, 7.0, 8.5,9.0, 10.5, 11.0, 12.5, 13.0, 14.5, 15.0, 16.5, 18.0, 19.5],
    "Horas_sueño": [5.0, 4.5, 6.0, 5.5, 7.0, 6.5, 5.0, 8.0, 7.5, 6.0,8.5, 7.0, 6.5, 8.0, 7.5, 8.5, 9.0, 8.0, 7.5, 8.0],
    "Ejercicio": [0.5, 0.0, 1.0, 0.5, 1.5, 1.0, 0.0, 2.0, 1.5, 1.0,2.5, 2.0, 1.5, 2.0, 1.0, 2.5, 2.0, 2.5, 3.0, 2.0],
    "Calificacion": [45, 50, 48, 55, 60, 58, 65, 70, 68, 75,72, 80, 85, 82, 88, 90, 92, 95, 98, 100]
}


df2 = pd.DataFrame(data2)
df2
````

Actividades
1. Define:
* X = variables independientes (todas excepto Calificacion)
* y = Calificacion
2. Divide en entrenamiento y prueba
3. Entrenamieto y predicción
4. Muestra:
* Coeficientes del modelo
* Intercepto
* Escribe el modelo: La fórmula con base en los coeficientes y el valor del intercepto.
5. Evalúa con:
* MAE
* MSE
* R²
6. Genera una predicción (y) proponiendo valores: Horas_estudio:x1 /Horas_sueño:x2/Ejercicio:x3
7. Interpretación general del modelo
* a.  Intepreta los resutlados de la evaluación MSE y R2
* b. ¿Qué variable influye más y cuál menos? (fíjate en los valores de los coeficientes obtenidos)
* c. ¿Como es su desempeño con respecto al MODELO DE REGRESIÓN SIMPLE? ¿Qué modelo escogerías para este caso-simple o el múltiple? ¿Por Qué?

````python
X = df2[['Horas_estudio', 'Horas_sueño', 'Ejercicio']]
y = df2['Calificacion']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)

from sklearn.linear_model import LinearRegression
modelo_mult = LinearRegression()
modelo_mult.fit(X_train, y_train)
y_pred = modelo_mult.predict(X_test)

print("--- Parámetros del Modelo ---")
print(f"Intercepto (b): {modelo_mult.intercept_:.4f}")

coef_df = pd.DataFrame(modelo_mult.coef_, X.columns, columns=['Coeficiente'])
print(coef_df)

from sklearn import metrics
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
r2 = metrics.r2_score(y_test, y_pred)

print("\n--- Evaluación del Modelo ---")
print(f"MAE (Error Absoluto Medio): {mae:.4f}")
print(f"MSE (Error Cuadrático Medio): {mse:.4f}")
print(f"R² (Coeficiente de Determinación): {r2:.4f}")
````

4. Escribe el modelo a partir de los coeficientes y el intercepto:

$y = ([Valor de Horas\_estudio] * Horas\_estudio) + ([Valor de Horas\_sueño] * Horas\_sueño) + ([Valor de Ejercicio] * Ejercicio) + [Valor del Intercepto]$

6. Prediccción ejemplo:



````python
nuevos_datos = np.array([[10, 7, 1]])

prediccion = modelo_mult.predict(nuevos_datos)

print(f"La calificación predicha para los valores ingresados es: {prediccion[0]:.2f}")
````

**7. Interpretación general del modelo**

a.
El R² nos dice qué tanto entiende el modelo la realidad si el valor es cercano a 1 , significa que el modelo explica casi todo lo que pasa con las calificaciones. El MSE es el margen de error entre más pequeño sea ese número, más exactas son las predicciones del modelo.

b.
La variable que más influye es la que tiene el número más alto en los coeficientes, porque un pequeño cambio ahí sube mucho la calificacion. La que menos influye es la que tiene el número más bajo, ya que afecta menos al resultado final.

c.
El Modelo Múltiple es mejor.
 Porque en la vida real la calificación no depende de una sola cosa. Al incluir el sueño y el ejercicio, el modelo se vuelve más inteligente y acertado. El R² suele ser más alto en el múltiple, lo que demuestra que es más preciso para predecir la nota final.
